```python
#creating sublists of artists and albums uries
def split_list(input_list, chunk_size):
    it = iter(input_list)
    return[list(islice(it, chunk_size)) for _ in range(0, len(input_list), chunk_size)]
          
artists_sublist_size = 50
albums_sublist_size = 20

pulled_artists_uries = []
pulled_albums_uries = []

print(f"artist uri: {final_df["artist_uri"][0].split(":")[2]}")
print(f"album uri: {final_df["album_uri"][0].split(":")[2]}")

for uri in artistsURI:
    pulled_artists_uries.append(uri.split(":")[2])

for uri in albumURI:
    pulled_albums_uries.append(uri.split(":")[2])

artists_sublists = split_list(pulled_artists_uries, artists_sublist_size)
albums_sublists = split_list(pulled_albums_uries, albums_sublist_size)

#print(f"artists uries sublist: {artists_sublists[:2]}")
#print(f"albums uries sublist: {artists_sublists[:2]}")
#print(f"length of albums uries sublist: {len(albums_sublists[0])}")
#print(f"length of artists uries sublist: {len(artists_sublists[0])}")
```

```python
# 6000 requests

count = 0
min_chunk = 5918
max_chunk = 6000

for chunk in artists_sublists:

    if count < min_chunk:
        count += 1       
        continue
    if count >= max_chunk:
        break
    
    
    response_API = requests.get(f"https://api.spotify.com/v1/artists",
                                   params={"ids": ",".join(chunk)},
                            headers = {"Authorization" : f"Bearer {accessToken}"})

    chunk_data = ""
    try:
        chunk_data = response_API.json()
    except:
        print(f"Borked on chunk {count}")
        print(response_API.content)
        break
        
    chunk_artists = chunk_data["artists"]

    count += 1 
   
    
    with open(f"/Users/maksym.tseranu/projects/spotify/spotify_million_playlist_dataset/data/artists_chunk_{count}.csv", 'w') as csv_file:
        csv_writer = csv.writer(csv_file, delimiter = ";")
        csv_writer.writerow(["id", "artist_followers, artist_genres, artist_popularity"])

        for artist in chunk_artists:          
            artist_row = [artist["id"], artist["followers"]["total"], ",".join(artist["genres"]), artist["popularity"]]
            csv_writer.writerow(artist_row)
```
